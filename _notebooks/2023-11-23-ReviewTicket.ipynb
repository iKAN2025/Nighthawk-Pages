## Binary Badge Challenge

[Link to Project](https://jplip.github.io/frontTri2/challenge/)


![challenge.png](/Nighthawk-Pages/images/challenge.png)

Explanation of Code: 

### Logic Gate Functions
- `andGate(input1, input2)` and `orGate(input1, input2)` simulate the AND and OR logic gates.
- They take two binary inputs and return the result based on the gate's logic.

### `generateLogicGateChallenge()` Function
- Dynamically generates challenge items for each logic gate (AND and OR).
- Creates elements for labels, input fields for binary values, and a "Check Answer" button for each gate.
- The event listener attached to the button handles user input, checks it against the expected gate operation, and provides feedback on correctness or prompts for valid inputs.

### User Interaction and Challenges
- For each logic gate (AND and OR), challenge items are created consisting of:
  - A label indicating the type of gate operation (AND or OR).
  - Two input fields for users to input binary values (0 or 1).
  - A "Check Answer" button to verify the user's input against the logic gate operation.
- Upon clicking the "Check Answer" button:
  - It validates user inputs, checks if they're valid binary values, and compares them against the expected output of the corresponding logic gate.
  - If the user's inputs match the expected gate operation, it provides an alert indicating correctness. Otherwise, it alerts the user to try again.

### Potential Enhancements
- Progress tracking or badge awards can be added upon correct answers.
- Styling enhancements or additional gate types can be included to expand the challenge.

This code provides a basic implementation of a binary logic gates challenge where users interact by inputting binary values and verifying their correctness against predefined logic gate operations (AND and OR).




## Binary Music
[Link to Project](https://jplip.github.io/frontTri2/music/)
![challenge.png](/Nighthawk-Pages/images/music.png)

### User Instructions
- A `<div>` with `id="instructions"` provides user instructions:
  - It includes headings and paragraphs explaining how the tool works.
  - Instructions guide users on inputting binary patterns and understanding the representation of '0's and '1's.

### Input Field and Button
- An `<input>` field with `type="text"` and an associated label allows users to input binary patterns.
- A `<button>` with an `onclick` event triggers the `composeMusic()` function to generate the musical composition.

### JavaScript Functions
- `composeMusic()`:
  - Triggered by the "Compose Music" button, it generates a musical composition based on the entered binary pattern.
- `noteToFrequency(note)`:
  - Maps musical notes ('C4', 'D4', etc.) to their respective frequencies using a predefined `notesMap`.
  - Returns the frequency of a given note.

### Audio Generation
- Uses the Web Audio API to generate sound.
- Creates an `AudioContext` and an `Oscillator` for sound wave production.
- For each character (bit) in the binary sequence:
  - If '0', sets the oscillator frequency with `noteToFrequency()` and creates a short note.
  - If '1', introduces a brief silence by advancing time without playing notes.

### Note Duration and Delay
- `noteDuration` and `noteDelay` variables control the duration of each note and delay between notes.

This code interprets a binary sequence input with '0's as musical notes and '1's as brief silences, generating a musical composition accordingly.
